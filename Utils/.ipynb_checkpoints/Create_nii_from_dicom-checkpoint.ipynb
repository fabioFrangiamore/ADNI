```python
import os, subprocess, time
```

```python
os.listdir()
```

```python
subject = []
root_dir = []
for dirs in os.listdir():
    for root, dirs, files in os.walk(dirs):
        last_dir = root.split('/')[-1]
        if last_dir[0] == 'S' and 'SAG' not in last_dir.upper() and 'Field_Mapping' not in root:
            print(root)
            print("++++++++++++")
            subject.append(last_dir)
            root_dir.append(root)
```

```python
dcm2niix_ = 'dcm2niix/build/bin/dcm2niix'
```

```python
dcm2niix_dir = '/'.join([scaricati_dir,'Scaricati', dcm2niix_])
```

```python
for i, dirs in enumerate(root_dir):
    print(dirs)
    print("++++++++++++++++++")
    print("\n"+ str(i) + '/' + str(len(root_dir)))
    command = [dcm2niix_dir, '/'.join([os.getcwd(), dirs])]
    print(command)
    registration = subprocess.Popen(command, stdout=subprocess.PIPE)
    output, errors = registration.communicate()
    print ([registration.returncode, errors, output])
    time.sleep(3)
```
